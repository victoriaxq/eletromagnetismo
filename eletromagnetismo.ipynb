## **Discretização do Domínio**
  - A discretização do domínio envolve dividir a região de interesse em pequenos elementos onde as equações diferenciais podem ser aproximadas. Aqui, estamos trabalhando com uma stripline em um domínio bidimensional $0 ≤ x ≤ a$ e $0 ≤ y ≤ b$.
  - Para aplicar o método dos elementos finitos, o domínio contínuo deve ser discretizado em pequenos elementos finitos. No caso de um domínio retangular, a discretização é feita dividindo o retângulo em $nx$ subdivisões ao longo da direção $x$ e $ny$ subdivisões ao longo da direção $y$. Cada subdivisão retangular é então dividida em dois triângulos.
  - Os nós da malha são os pontos de interseção das subdivisões. As coordenadas dos nós são geradas uniformemente ao longo do retângulo. Para um retângulo de dimensões $a x b$, com nx subdivisões na direção $x$ e $ny$ na direção $y$, as coordenadas dos nós $(Xi,Yj)$ são:
$X_i = \frac{i \cdot a}{nx}, \quad Y_j = \frac{j \cdot b}{ny}, \text{ onde } i = 0, 1, \ldots, nx \text{ e } j = 0, 1, \ldots, ny$
  - Cada retângulo formado pelas subdivisões é dividido em dois triângulos. A conectividade dos triângulos é determinada pelos índices dos nós:
    - Para cada subdivisão retangular, com nós $n1, n2, n3$ e $n4$:
      - Triângulo inferior esquerdo: $(n1,n2,n3)$
      - Triângulo superior direito: $(n2,n4,n3)$

## **Sistema Linear para um Único Elemento**
Para obter o sistema linear para um único elemento, deve-se considerar a discretização do domínio contínuo em elementos menores, que são definidos como triângulos. A partir disso, serão definidas funções de forma associadas aos nós de cada triângulo, para aproximar o potencial elétrico dentro de cada um deles.

- **Funções de Interpolação**

  Dentro de um triângulo com vértices $(x_i, y_i)$, $(x_j, y_j)$ e $(x_k, y_k)$, pode-se escrever o potencial elétrico $V(x, y)$ como:

  $$
  V^e(x, y) \approx N_i^e(x, y) V_i + N_j^e(x, y) V_j + N_k^e(x, y) V_k
  $$

  onde:
  - $V_i, V_j$ e $V_k$ são os potenciais nos vértices do triângulo.
  - $N_i^e, N_j^e$ e $N_k^e$ são as funções de forma associadas a cada nó do triângulo.

- **Definição das Funções de Forma**

  As funções de forma $N_i^e, N_j^e$ e $N_k^e$ para um elemento "e" são definidas como:

  $$
  N_i^e(x, y) = a_i + b_i x + c_i y
  $$

- **Cálculo dos Coeficientes das Funções de Forma**

  Os coeficientes $a_i, b_i, c_i$ são determinados resolvendo o seguinte sistema de equações:

  $$
  \begin{bmatrix}
  1 & x_i & y_i \\
  1 & x_j & y_j \\
  1 & x_k & y_k
  \end{bmatrix}
  \begin{bmatrix}
  a_i \\
  b_i \\
  c_i
  \end{bmatrix}
  =
  \begin{bmatrix}
  1 \\
  0 \\
  0
  \end{bmatrix}, \quad
  \begin{bmatrix}
  1 & x_i & y_i \\
  1 & x_j & y_j \\
  1 & x_k & y_k
  \end{bmatrix}
  \begin{bmatrix}
  a_j \\
  b_j \\
  c_j
  \end{bmatrix}
  =
  \begin{bmatrix}
  0 \\
  1 \\
  0
  \end{bmatrix}, \quad
  \begin{bmatrix}
  1 & x_i & y_i \\
  1 & x_j & y_j \\
  1 & x_k & y_k
  \end{bmatrix}
  \begin{bmatrix}
  a_k \\
  b_k \\
  c_k
  \end{bmatrix}
  =
  \begin{bmatrix}
  0 \\
  0 \\
  1
  \end{bmatrix}
  $$

- **Determinação da Área do Triângulo**

  A área do triângulo é calculada utilizando a fórmula:

  $$
  A_e = \frac{1}{2} \left| x_i (y_j - y_k) + x_j (y_k - y_i) + x_k (y_i - y_j) \right|
  $$

- **Funções de Forma Explícitas**

  Os coeficientes das funções de forma são:

  $$
  a_i = \frac{1}{2A_e} (x_j y_k - x_k y_j), \quad b_i = \frac{1}{2A_e} (y_j - y_k), \quad c_i = \frac{1}{2A_e} (x_k - x_j)
  $$

  $$
  a_j = \frac{1}{2A_e} (x_k y_i - x_i y_k), \quad b_j = \frac{1}{2A_e} (y_k - y_i), \quad c_j = \frac{1}{2A_e} (x_i - x_k)
  $$

  $$
  a_k = \frac{1}{2A_e} (x_i y_j - x_j y_i), \quad b_k = \frac{1}{2A_e} (y_i - y_j), \quad c_k = \frac{1}{2A_e} (x_j - x_i)
  $$

  Portanto, as funções de forma explícitas são:

  $$
  N_i^e(x, y) = \frac{1}{2A_e} [(x_j y_k - x_k y_j) + (y_j - y_k)x + (x_k - x_j)y]
  $$

  $$
  N_j^e(x, y) = \frac{1}{2A_e} [(x_k y_i - x_i y_k) + (y_k - y_i)x + (x_i - x_k)y]
  $$

  $$
  N_k^e(x, y) = \frac{1}{2A_e} [(x_i y_j - x_j y_i) + (y_i - y_j)x + (x_j - x_i)y]
  $$

- **Matriz de Rigidez Local**

  A matriz de rigidez local $K_e$ para um elemento triangular é dada por:

  $$
  K_e = \frac{1}{4A_e}
  \begin{bmatrix}
  b_i^2 + c_i^2 & b_i b_j + c_i c_j & b_i b_k + c_i c_k \\
  b_j b_i + c_j c_i & b_j^2 + c_j^2 & b_j b_k + c_j c_k \\
  b_k b_i + c_k c_i & b_k b_j + c_k c_j & b_k^2 + c_k^2
  \end{bmatrix}
  $$

- **Vetor de Força Local**

  O vetor de força local $F_e$ é dado por:

  $$
  F_e = \frac{A_e}{3}
  \begin{bmatrix}
  f(x_i, y_i) \\
  f(x_j, y_j) \\
  f(x_k, y_k)
  \end{bmatrix}
  $$

- **Sistema Linear para o Elemento**

  Com a matriz de rigidez local $K_e$ e o vetor de força local $F_e$, o sistema linear para o elemento é:

  $$
  K_e \mathbf{V}_e = F_e
  $$

  onde:

  $$
  K_e = \frac{1}{4A_e}
  \begin{bmatrix}
  b_i^2 + c_i^2 & b_i b_j + c_i c_j & b_i b_k + c_i c_k \\
  b_j b_i + c_j c_i & b_j^2 + c_j^2 & b_j b_k + c_j c_k \\
  b_k b_i + c_k c_i & b_k b_j + c_k c_j & b_k^2 + c_k^2
  \end{bmatrix}, \quad
  \mathbf{V}_e = \begin{bmatrix}
  V_i \\
  V_j \\
  V_k
  \end{bmatrix}, \quad
  F_e = \frac{A_e}{3}
  \begin{bmatrix}
  f(x_i, y_i) \\
  f(x_j, y_j) \\
  f(x_k, y_k)
  \end{bmatrix}
  $$



# **2ª Questão**
## **Montagem do Sitema Global**

Para que o sistema global possa ser montado, é necessário compô-lo a partir dos elementos unitários vistos na primeira questão.

- **Matriz Global de Rigidez $K$**

    Tem-se a matriz global de rigidez somando-se as contribuições de cada elemento finito. Se um elemento tem nós $i$, $j$, $k$, suas partes são adicionadas às posições correspondentes na matriz global:

  $$
  \begin{align*}
  \mathbf{K}[i, i] &+= \mathbf{K}_e[0, 0], \\
  \mathbf{K}[i, j] &+= \mathbf{K}_e[0, 1], \\
  \mathbf{K}[i, k] &+= \mathbf{K}_e[0, 2], \\
  \mathbf{K}[j, i] &+= \mathbf{K}_e[1, 0], \\
  \mathbf{K}[j, j] &+= \mathbf{K}_e[1, 1], \\
  \mathbf{K}[j, k] &+= \mathbf{K}_e[1, 2], \\
  \mathbf{K}[k, i] &+= \mathbf{K}_e[2, 0], \\
  \mathbf{K}[k, j] &+= \mathbf{K}_e[2, 1], \\
  \mathbf{K}[k, k] &+= \mathbf{K}_e[2, 2]
  \end{align*}
  $$

- **Vetor Global de Força $\mathbf{F}$:**

  De maneira análoga ao que foi feito acima, o vetor global é formado ao somar as contribuições dos vetores locais:

$$
\begin{align*}
\mathbf{F}[i] &+= \mathbf{F}_e[0], \\
\mathbf{F}[j] &+= \mathbf{F}_e[1], \\
\mathbf{F}[k] &+= \mathbf{F}_e[2]
\end{align*}
$$

-  **Aplicação das Condições de Contorno**

    As condições de contorno são aplicadas para garantir que os nós nas fronteiras tenham os valores corretos.

  - **Nós na Stripline:**

      A stripline é mantida a um potencial $V_0$. Para esses nós, ajustamos a matriz global e o vetor global para fixar os valores, onde $n$ é o índice do nó na stripline.:

  $$
  \mathbf{K}[n, :] = 0, \quad \mathbf{K}[n, n] = 1, \quad \mathbf{F}[n] = V_0
  $$

  - **Nós nas Paredes Aterradas:**
    As paredes são aterradas, ou seja, têm potencial zero. Para esses nós, faz-se assim, lembrando que $m$ é o índice do nó na parede aterrada.:

  $$
  \mathbf{K}[m, :] = 0, \quad \mathbf{K}[m, m] = 1, \quad \mathbf{F}[m] = 0
  $$

- **Resolução do Sistema Linear**

    Após montar o sistema global e aplicar as condições de contorno, o sistema linear resultante é assim (em que $\mathbf{\Phi}$ é o vetor de potenciais nos nós da malha.):

$$
\mathbf{K} \cdot \mathbf{\Phi} = \mathbf{F}
$$


# **3ª e 4ª Questões**
## **Implementação do Código**

- O código utiliza o Método dos Elementos Finitos para resolver o problema do potencial entre placas em uma região retangular de 10 mm por 4 mm.

- Inicialmente, é gerada uma malha triangular com 10 subdivisões ao longo do eixo $x$ e 8 subdivisões ao longo do eixo $y$, totalizando $N = 160$ triângulos.

- Em seguida, são calculadas as matrizes de rigidez locais para cada triângulo e montada a matriz global de rigidez.

- Após isso, as condições de contorno são aplicadas, colocando o valor do potencial nas bordas do retângulo como igual a zero e definindo uma faixa metálica central de 2 mm com potencial  $V_0 = 1V$.

- Assim, o sistema linear resultante é resolvido para obter a distribuição do potencial em toda a região, e os resultados são visualizados em gráficos que mostram a malha e a distribuição do potencial.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import lil_matrix
from scipy.sparse.linalg import spsolve
from scipy.interpolate import griddata

# Dimensões do retângulo
a = 10  # mm
b = 4   # mm

# Parâmetros do problema
W = 2   # mm
Vo = 1  # V
epsilon_r = 1  # Permissividade relativa

# Número de subdivisões (múltiplo de 5 para x, par para y)
nx = 10
ny = 8
N = nx * ny * 2

def generate_mesh(a, b, nx, ny):
    # Coordenadas dos nós
    x = np.linspace(0, a, nx+1)
    y = np.linspace(0, b, ny+1)
    X, Y = np.meshgrid(x, y)
    X = X.flatten()
    Y = Y.flatten()

    # Conectividade dos triângulos
    triangles = []
    for j in range(ny):
        for i in range(nx):
            # Índices dos nós
            n1 = i + j*(nx+1)
            n2 = n1 + 1
            n3 = n1 + (nx+1)
            n4 = n3 + 1

            # Triângulo inferior esquerdo
            triangles.append([n1, n2, n3])
            # Triângulo superior direito
            triangles.append([n2, n4, n3])

    triangles = np.array(triangles)
    return x, y, X, Y, triangles

# Gerar a malha
x, y, X, Y, triangles = generate_mesh(a, b, nx, ny)

# Plotar a malha de triângulos
plt.figure(figsize=(8, 5))
plt.triplot(X, Y, triangles, '--', color='magenta', alpha=0.7, lw=1)
plt.plot(X, Y, 'o', color='magenta', markersize=5)
plt.xlabel('a')
plt.ylabel('b')
plt.title('Malha de Triângulos')
plt.show()
```

```python
# Função para montar a matriz de rigidez local de um triângulo
def local_stiffness_matrix(x1, y1, x2, y2, x3, y3, epsilon_r):
    A = 0.5 * np.abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2))
    b = np.array([y2 - y3, y3 - y1, y1 - y2])
    c = np.array([x3 - x2, x1 - x3, x2 - x1])
    B = np.array([b, c])
    return epsilon_r * (1.0 / (4.0 * A)) * (B.T @ B)

# Montagem da matriz global de rigidez
n_nodes = (nx + 1) * (ny + 1)
K = np.zeros((n_nodes, n_nodes))
F = np.zeros(n_nodes)

for tri in triangles:
    n1, n2, n3 = tri
    x1, y1 = X[n1], Y[n1]
    x2, y2 = X[n2], Y[n2]
    x3, y3 = X[n3], Y[n3]
    K_local = local_stiffness_matrix(x1, y1, x2, y2, x3, y3, epsilon_r)
    for i, ni in enumerate(tri):
        for j, nj in enumerate(tri):
            K[ni, nj] += K_local[i, j]

# Aplicação das condições de contorno
# Placas aterradas
boundary_nodes = np.where((Y == 0) | (Y == b) | (X == 0) | (X == a))[0]
for node in boundary_nodes:
    K[node, :] = 0
    K[node, node] = 1
    F[node] = 0

# Fita metálica no centro com potencial Vo
strip_nodes = np.where((X >= (a/2 - W/2)) & (X <= (a/2 + W/2)) & (Y == b/2))[0]
for node in strip_nodes:
    K[node, :] = 0
    K[node, node] = 1
    F[node] = Vo

# Resolução do sistema linea
V = np.linalg.solve(K, F)

# Plotagem do potencial usando a malha de triângulos
plt.figure(figsize=(9, 6))
plt.tricontourf(X, Y, triangles, V, cmap='magma')
plt.colorbar(label='Potencial (V)')
plt.xlabel('a')
plt.ylabel('b')
plt.title(f'Distribuição do Potencial para N = {N}')
plt.show()
```

# **5ª Questão**
## **Análise do Problema variando $N$**

Aqui, usou-se os parâmetros análogos aos da 4ª questão, modificando apenas o $N$. Assim, utilizou-se os seguintes valores de $N$:
*   $N_x=10$, $N_y=2$
*   $N_x=15$, $N_y=10$
*   $N_x=50$, $N_y=20$
*   $N_x=100$, $N_y=100$


```python
# Montagem da matriz global de rigidez
def solve_potential(a, b, W, Vo, epsilon_r, nx, ny):
    x, y, X, Y, triangles = generate_mesh(a, b, nx, ny)

    n_nodes = (nx + 1) * (ny + 1)
    K = np.zeros((n_nodes, n_nodes))
    F = np.zeros(n_nodes)

    for tri in triangles:
        n1, n2, n3 = tri
        x1, y1 = X[n1], Y[n1]
        x2, y2 = X[n2], Y[n2]
        x3, y3 = X[n3], Y[n3]
        K_local = local_stiffness_matrix(x1, y1, x2, y2, x3, y3, epsilon_r)
        for i, ni in enumerate(tri):
            for j, nj in enumerate(tri):
                K[ni, nj] += K_local[i, j]

    boundary_nodes = np.where((Y == 0) | (Y == b) | (X == 0) | (X == a))[0]
    for node in boundary_nodes:
        K[node, :] = 0
        K[node, node] = 1
        F[node] = 0

    strip_nodes = np.where((X >= (a/2 - W/2)) & (X <= (a/2 + W/2)) & (Y == b/2))[0]
    for node in strip_nodes:
        K[node, :] = 0
        K[node, node] = 1
        F[node] = Vo

    V = np.linalg.solve(K, F)
    return X, Y, triangles, V

# Função para plotar o potencial
def plot_potential(a, b, nx_values, ny_values, W, Vo, epsilon_r):
    plt.figure(figsize=(14, 10))
    for idx, (nx, ny) in enumerate(zip(nx_values, ny_values)):
        X, Y, triangles, V = solve_potential(a, b, W, Vo, epsilon_r, nx, ny)
        N = nx * ny * 2

        plt.subplot(2, len(nx_values)//2, idx + 1)
        plt.tricontourf(X, Y, triangles, V, cmap='magma')
        plt.colorbar(label='Potencial (V)')
        plt.xlabel('a')
        plt.ylabel('b')
        plt.title(f'Distribuição do Potencial para N = {N}')
    plt.tight_layout()
    plt.show()

# Valores de N para testar
nx_values = [10, 15, 25, 100]  # Múltiplos de 5
ny_values = [2, 10, 20, 100]   # Valores pares

plot_potential(a, b, nx_values, ny_values, W, Vo, epsilon_r)
```
